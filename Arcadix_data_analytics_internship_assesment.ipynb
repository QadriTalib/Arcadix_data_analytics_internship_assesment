```python
# Box plot for Monthly Income
ax = sns.boxplot(x=df["MonthlyIncome"])
plt.show(ax)
```

```python
ay = sns.boxplot(x = "Gender", y = "MonthlyIncome", data=df)
plt.show(ay)
```

```python
ay = sns.boxplot(x = "Gender", y = "MonthlyIncome", hue = "JobRole", data=df, palette="Set3")
plt.show(ay)
```

```python
# Creating histogram for distribution of Age
y = df.TotalWorkingYears
n_bins = 6
N, bins, patches = plt.hist(y, bins = n_bins, facecolor = 'Blue')
plt.show()
```

```python
fig = px.pie(df, values = df["MonthlyRate"], names = df["JobRole"],
             title = "MonthlyRate Based on JobRole", hole = .5)
fig.update_traces(textposition="inside", textinfo="percent + label")
fig.show()
```
